```python
import pyperclip

from collections import Counter

import numpy as np

import pandas as pd

from bokeh.plotting import figure
from bokeh.io import output_notebook, show
import holoviews as hv
from holoviews import opts
hv.extension('bokeh')

from scipy.cluster.hierarchy import dendrogram, linkage, ward
```

```python
from Bio import SeqIO

# Specify the path to your FASTA file
fasta_file_path = "data/ribosomalRNA.fasta"

# Use SeqIO.parse to read the FASTA file and store the sequences in a list
sequences = list(SeqIO.parse(fasta_file_path, "fasta"))

# create a for loop to print out the sequences and their IDs

for sequence in sequences:
    print(sequence.id,"\n",sequence.seq)
```

```python
def count_nucleotide (organismsbases):
    '''
    '''
    # organismsbases = organismsbases.replace('\n', '')
    count_base = Counter(organismsbases)
    most_comm_base = count_base.most_common(2)
    print("The three results are the number of each nucleotide, sum of all nucleotides and the 2 most common bases found.")
    return count_base, sum(count_base.values()),most_comm_base

for sequence in sequences:
    print(sequence.id, count_nucleotide(sequence.seq))
```

```python
def gc_content(seq):
    gc = (seq.count("C") + seq.count("G")) / (len(seq)) * 100
    
    return gc

def ga_content(seq):

    ga = (seq.count("G") + seq.count("A")) / len(seq) * 100

    return ga

for sequence in sequences:
    gc = gc_content(sequence.seq)
    print(sequence.id, gc)

df2.insert(loc=4, column="GC_content", value=[gc_content(sequence.seq) for sequence in sequences])

df2.insert(loc=5, column="GA_content", value=[ga_content(sequence.seq) for sequence in sequences])
```

```python
# for clustering
content_nuc = df2[["GC_content","GA_content"]]

print ("")
print(type(content_nuc))
print ("")
print (content_nuc)
```

```python
import matplotlib.pyplot as plt

samp_orgs = df2.drop(["GC_content", "GA_content", "organism"], axis = 1).values

merge = linkage(samp_orgs, method = "complete")

merge2 = ward(samp_orgs)
dendrogram(merge, labels = ["E_coli", "M_tuberculosis", "N_gonorrhoeae", "Mycobacterium", "P_multocida", "Salmonella", "M_smegmatis", "N_meningitidis", "C_psittaci", "B_burgdorferi", "B_burgdorferiBol26", "M_abscessus", "C_acnes", "E_coli_K12", "M_smegmatis", "M_smegmatis_MC2_155", "H_salinarium", "H_pylori", "E_coli_str_K12", "C_reinhardtii"], leaf_rotation = 90, leaf_font_size = 20)

plt.xlabel("Organisms")
plt.ylabel("Cluster Distance")
plt.show()
```
